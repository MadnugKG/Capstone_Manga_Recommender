### Import librairies

```python
import pandas as pd
import os
from os.path import join
```

### Define Functions

```python
def compile_files(path):
    count = 0
    compiled_file = pd.DataFrame(list())
    for root, dirs, files in os.walk(path):
        if len(files) > 0:
            for file in files:
                temp = pd.read_csv(join(root, file))
                temp.drop(columns=['Unnamed: 0'], inplace=True)
                temp['file_index'] = file.split("_")[-1][:-4]
                print(f"{join(root, file)}: Start")
                if compiled_file.shape[0]:
                    compiled_file = pd.concat([compiled_file, temp])
                else:
                    compiled_file = temp
                print(f"{join(root, file)}: Completed")

    return compiled_file
        
```
